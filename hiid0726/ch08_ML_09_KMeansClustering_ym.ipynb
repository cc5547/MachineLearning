```python
# 주요 라이브러리 import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# https://github.com/BigData23th/Data/raw/main/example_cluster.csv
df = pd.read_csv('https://github.com/BigData23th/Data/raw/main/example_cluster.csv')
df
```

```python
# 산점도
sns.scatterplot(x='var_1', y='var_2', data=df)
plt.show()
```

```python
# K 평균 군집화 (K Means Clustering)
from sklearn.cluster import KMeans
```

```python
# n_clusters : 나눠줄 클러스터(묶음)의 수
kmeans_model = KMeans(n_clusters=3, random_state = 2023)
```

```python
# 지도 -> X, y
# 비지도 -> X
kmeans_model.fit(df)
```

```python
# 예측
# 군집화 분석을 통해 분석된 K개의 그룹과 각 데이터가 어느 그룹에 속했는지
# fit_predict (fit하고 바로 예측)
kmeans_model.predict(df)
pred = kmeans_model.fit_predict(df)
```

```python
# label
# KMeans 산출한 그룹들을 label로 주입
df['label'] = pred
```

```python
# 산점도2 - label을 적용한
# sns.scatterplot(
#     x='var_1', y='var_2', data=df,
#     hue='label'
# )
sns.scatterplot(
    x='var_1', y='var_2', data=df,
    hue='label',
    palette = 'rainbow'
)
```

```python
# 이너셔 (inertia)
# - 각 그룹에서의 중심과 각 그룹에 해당하는 데이터 간의 거리에 대한 합
kmeans_model.inertia_
```

```python
# k가 엄청 많을 경우
kmeans_model2 = KMeans(n_clusters=500, random_state = 2023)
kmeans_model2.fit(df)
kmeans_model2.inertia_
```

```python
df['label2'] = kmeans_model2.predict(df)
```

```python
sns.scatterplot(
    x='var_1', y='var_2', data=df,
    hue='label2',
    palette = 'rainbow'
)
```

```python
# 적절한 k값 설정 -> 엘보우 기법
df = df.drop(['label', 'label2', 'lable2'], axis = 1)
df.columns
```

```python
# inertia 거리의 변화폭이 둔화되는 지점 -> 꺾이는 지점
distance=[]
for k in range(2,10) : #K = 2~9
  k_model = KMeans(n_clusters=k, random_state = 2023)
  k_model.fit(df)# fit만 시켜도 inertia_ 생성
  distance.append(k_model.inertia_)
```

```python
distance
```

```python
# 선 차트
sns.lineplot(x=range(2,10), y=distance)
# 엘보우 : 직관적이고 편리하긴 하나, 객관적 건 아님 -> 사람이 판단
```

```python
for i in range(1, len(distance)):
    # 기울기 -> x 증가 -> y가 증가
    difference = abs(distance[i] - distance[i-1])
    print(f"{i+1}->{i+2}", difference)
```

```python
# 이상거래 데이터 -> 단순화
# https://github.com/BigData23th/Data/raw/main/customer.csv
df = pd.read_csv('https://github.com/BigData23th/Data/raw/main/customer.csv')
```

```python
# cc_num : credit card -> 카드번호 -> 특정 고객
df.cc_num.nunique()
```

```python
# category별로 고객들이 얼마나 소비를 하냐
customer = pd.get_dummies(df, columns=['category'])
customer.info()
```

```python
# iloc -> 정수 인덱스로 행, 열을 가져오겠다 (인덱서)
# : -> 전체, [행, 열] => 전체 행 중에 2번째 이후 (경계 포함) 열들만 표시하겠다
customer.iloc[:,2:]
```

```python
# 특정 카테고리 포함 여부 (1, 0)
for col_name in customer.iloc[:,2:].columns: # 범주들 이름으로 for문 (반복)
    # 트랜잭션, 거래, 행 (데이터에서의 행)
    # print(col_name)
    # 기존에 있는 컬럼(열)에 새로운 값을 연산
    customer[col_name] = customer[col_name] * customer['amt']
    # 특정 범주(카테고리 포함 여부) : customer[col_name] : 1, 0
    # 소비액 : customer['amt']
    # -> 특정 카테고리에 대한 소비액
    # 한 행 -> 거래 => 열 -> 거래가 속한 범주의 열이라면 amt랑 곱해진 소비액 나올 거고,
    # 나머지 소속되지 않은 경우에 -> 0만 있을 것
```

```python
customer
```

```python
# cc_num 별로 합계를 해주겠다
# 카드번호별로 groupby -> sum()
customer_agg = customer.groupby('cc_num').sum()
customer_agg
# cc_num별 총 소비액인 amt
# 각기 category별 cc_num의 소비액인 각자 열
# -> 총 소비 또는 각 범주별 소비를 통해서 '분류', '군집화'가 가능
```

```python
customer_agg.info()
```

```python
# 스케일러 -> 표준화 스케일러 (Standard Scaler)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# 스케일링을 하고 나서 데이터
scaled = scaler.fit_transform(customer_agg)
scaled # numpy array
```

```python
scaled_df = pd.DataFrame(scaled, columns = customer_agg.columns, index = customer_agg.index)
```

```python
scaled_df
```

```python
# 엘보우 기법
distance = []
for k in range(2, 10): # 끝경계 포함 X
    k_model = KMeans(n_clusters=k, random_state = 2023)
    k_model.fit(scaled_df)
    distance.append(k_model.inertia_)
```

```python
for i in range(1, len(distance)):
    # 기울기 -> x 증가 -> y가 증가
    difference = abs(distance[i] - distance[i-1])
    print(f"{i+1}->{i+2}", difference)
```

```python
sns.lineplot(x=range(2,10), y=distance)
```

```python
# 실루엣 계수
# - 이너셔(inertia) : 그룹(클러스터)의 중심으로부터 다른 요소들간의 거리
# - 실루엣 계수 : 클러스터 내부의 평균 거리 + 최근접한 다른 클러스터와의 평균 거리
from sklearn.metrics import silhouette_score
```

```python
silhouette = []
for k in range(2, 10):
    k_model = KMeans(n_clusters=k, random_state = 2023)
    # 실제의 데이터 값과, 분류된 라벨을 통해서 클러스터 내부 및 클러스터 간 거리를 계산
    labels = k_model.fit_predict(scaled_df)
    silhouette.append(silhouette_score(scaled_df,labels))
```

```python
silhouette
```

```python
# 실루엣 계수는 더 클 수록 좋은 분류를 하고 있는 것
sns.lineplot(x=range(2,10), y=silhouette)
```

```python
k_model = KMeans(n_clusters=4, random_state = 2023)
labels = k_model.fit_predict(scaled_df)
```

```python
scaled_df['label'] = labels
```

```python
scaled_df
```

```python
scaled_df_mean = scaled_df.groupby('label').mean()
scaled_df_mean
```

```python
scaled_df_count = scaled_df.groupby('label').count().amt
scaled_df_count
```

```python
scaled_df_count = scaled_df_count.rename('count')
scaled_df_all = scaled_df_mean.join(scaled_df_count)
```
